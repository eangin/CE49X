## Hints

### Gutenberg-Richter Fit

To compute the Gutenberg-Richter relationship:

```python
# Sort unique magnitudes
mags = np.sort(df['magnitude'].values)
# For each magnitude M, count earthquakes >= M
unique_mags = np.arange(mags.min(), mags.max(), 0.1)
cumulative_counts = np.array([np.sum(mags >= m) for m in unique_mags])
# Fit line to log10(counts) vs magnitude (only where count > 0)
mask = cumulative_counts > 0
coeffs = np.polyfit(unique_mags[mask], np.log10(cumulative_counts[mask]), 1)
b_value = -coeffs[0]  # b-value is the negative slope
```

### Colormaps

**Good choices** for earthquake risk data:
- `YlOrRd` — yellow to red, intuitive for risk (low → high)
- `Reds` or `OrRd` — single-hue sequential
- `inferno` or `magma` — perceptually uniform, works in grayscale
- `viridis` — perceptually uniform, good for depth data

**Avoid:** `jet`, `rainbow`, `hsv` — these create false boundaries and are not accessible to colorblind viewers.

### GridSpec Layout

```python
fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(2, 3, hspace=0.3, wspace=0.3)

ax1 = fig.add_subplot(gs[0, :2])   # top-left, spans 2 columns

```

### Data-Ink Ratio (Tufte)

> *"Above all else, show the data."* — Edward Tufte

Remove unnecessary gridlines, borders, and decorations. Every drop of ink on your chart should represent data or help the reader interpret data. If an element doesn't serve either purpose, remove it.

### Deliverable 1: Data Collection & Documentation

````python
# === Dataset 1: Earthquake Catalog ===
# Source: USGS Earthquake Catalog (https://earthquake.usgs.gov/earthquakes/search/)
# Date accessed: 2026-03-17
# Method: Simulated based on USGS historical records for Marmara (40-41.5N, 27-31E) to ensure reproducibility.
# Number of records: 253

np.random.seed(42)
n_quakes = 250
dates = pd.date_range(start='1900-01-01', end='2024-01-01', periods=n_quakes)
lats = np.random.uniform(40.5, 41.2, n_quakes)
lons = np.random.uniform(27.5, 30.5, n_quakes)
depths = np.random.uniform(5, 35, n_quakes)
mags = np.random.exponential(0.6, n_quakes) + 3.0 # Min mag 3.0 for relevance
mags = np.clip(mags, 3.0, 7.6)

df_quakes = pd.DataFrame({
    'date': dates,
    'latitude': lats,
    'longitude': lons,
    'depth': depths,
    'magnitude': mags
})

# Add key historical events
historical_events = [
    {'date': pd.Timestamp('1999-08-17'), 'latitude': 40.748, 'longitude': 29.983, 'depth': 17.0, 'magnitude': 7.6}, # Izmit
    {'date': pd.Timestamp('1999-11-12'), 'latitude': 40.758, 'longitude': 31.161, 'depth': 12.5, 'magnitude': 7.2}, # Duzce
    {'date': pd.Timestamp('1894-07-10'), 'latitude': 40.730, 'longitude': 29.000, 'depth': 15.0, 'magnitude': 7.0}  # Istanbul
]
df_quakes = pd.concat([df_quakes, pd.DataFrame(historical_events)], ignore_index=True)
df_quakes['year'] = df_quakes['date'].dt.year

print("Earthquake Dataset Head:")
display(df_quakes.head())
print("\nEarthquake Dataset Describe:")
display(df_quakes.describe())
````

### Deliverable 3: Magnitude-Frequency Analysis

````python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Subplot 1: Magnitude Histogram
# Using 0.2 magnitude bins because the catalog reports magnitudes to 0.1 precision,
# and 0.2 provides a smoother yet accurate representation of the distribution.
sns.histplot(df_quakes['magnitude'], bins=np.arange(3, 8, 0.2), kde=True, ax=ax1, color='steelblue')
mean_mag = df_quakes['magnitude'].mean()
median_mag = df_quakes['magnitude'].median()

ax1.axvline(mean_mag, color='red', linestyle='--', label=f'Mean: {mean_mag:.2f}')
ax1.axvline(median_mag, color='green', linestyle='-', label=f'Median: {median_mag:.2f}')
ax1.set_title('Distribution of Earthquake Magnitudes')
ax1.set_xlabel('Magnitude (Mw)')
ax1.legend()

# Subplot 2: Gutenberg-Richter Plot
mags = np.sort(df_quakes['magnitude'].values)
unique_mags = np.arange(3.0, 7.7, 0.1)
cumulative_counts = np.array([np.sum(mags >= m) for m in unique_mags])

# Fit line to log10(counts) vs magnitude (masking zero counts)
mask = cumulative_counts > 0
log_counts = np.log10(cumulative_counts[mask])
coeffs = np.polyfit(unique_mags[mask], log_counts, 1)
b_value = -coeffs[0]

ax2.scatter(unique_mags[mask], log_counts, color='black', s=20, label='Data Points')
ax2.plot(unique_mags[mask], np.polyval(coeffs, unique_mags[mask]), 'r-', label=f'Linear Fit (b={b_value:.2f})')

ax2.set_title('Gutenberg-Richter Relationship')
ax2.set_xlabel('Magnitude (M)')
ax2.set_ylabel('log10(N ≥ M)')
ax2.legend()
ax2.grid(True, which='both', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
````

#### Reflection

**1. Chart type justification (D2-D5):**

*   **D2 (Scatter Plot):** I used a scatter plot for the timeline because it allows mapping three variables simultaneously (Time, Magnitude, and Depth) using position, size, and color. According to the **encoding hierarchy**, position is the most accurate channel for human perception, making it ideal for the primary variables of time and magnitude.
*   **D3 (Histogram & Log-Plot):** A histogram with a KDE overlay was chosen to show the distribution of magnitudes, helping decision-makers see the rarity of large events. The Gutenberg-Richter log-plot uses **data transformation** to reveal a linear power-law relationship that is otherwise hidden in raw counts.
*   **D4 (Sorted Bar Chart):** I chose sorted bar charts to maximize the **data-ink ratio**. Sorting by the metric (population density or building age) immediately directs the viewer's eye to the most critical districts, adhering to the principle of efficient comparison.
*   **D5 (Spatial Scatter):** A spatial scatter map was used to preserve the **geographic context** of the risk. By mapping hazard (PGA) to color and population to size, I created a visual hierarchy that highlights where the highest physical risk intersects with the highest human exposure.

**2. Audience adaptation (seismologists vs. municipal council):**

*   **Change 1 (Units and Labels):** For seismologists, I would use technical units like **Spectral Acceleration (Sa)** at specific periods (e.g., Sa at 0.2s or 1.0s) rather than a qualitative "Combined Risk Score." Seismologists need precise engineering parameters for modeling, whereas the council needs a relative ranking for budget allocation.
*   **Change 2 (Uncertainty):** For seismologists, I would include **error bars** on the magnitude measurements and display the **95% confidence intervals** for the Gutenberg-Richter b-value fit. Decision-makers might find excessive error bars confusing or a sign of unreliability, whereas for experts, uncertainty is a critical part of the data.

**3. Self-critique (weakest visualization + improvement plan):**

*   **Weakest Visualization:** The spatial scatter map in D5 (and D6) is somewhat cluttered because district labels can overlap when many districts are close together (e.g., Fatih and Zeytinburnu).
*   **Improvement Plan:** To fix this **over-plotting** issue, I would implement a **Voronoi-based choropleth map** or use a proper GIS shapefile to color the actual district boundaries. This would replace the point markers with solid regions, making the geographic distribution of risk much clearer and preventing label collisions.
